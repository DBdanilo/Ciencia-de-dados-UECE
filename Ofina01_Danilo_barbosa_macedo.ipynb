O objetivo da empresa é prever a temperatura de amanhã com base nas temperaturas anteriores, para fornecer uma previsão mais precisa. Para isso, você decide usar uma rede neural do tipo LSTM (Long Short-Term Memory), que é adequada para lidar com dados sequenciais.

### Seu objetivo nesta oficina é construir um modelo de deep learning utilizando uma rede neural LSTM que possa prever a temperatura do dia seguinte com base nas temperaturas dos últimos 30 dias.

```python
#Importando as bibliotecas necessárias

import numpy as np

import pandas as pd

import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import LSTM, Dense

from sklearn.preprocessing import MinMaxScaler

from tensorflow.keras.layers import Input
```

### Gerando dados fictícios de temperaturas diárias para os últimos 100 dias

```python
np.random.seed(0)

data = np.sin(np.linspace(0, 20, 100)) * 10 + 25 + np.random.normal(0, 1, 100)

 

# Transformando os dados em um DataFrame para facilitar a manipulação

df = pd.DataFrame(data, columns=['Temperatura'])

 

# Exibindo as primeiras 5 linhas do DataFrame para visualizar os dados

print(df.head(100))
```

### visualização dos dados Gerados

```python
plt.plot(df)

plt.title('Temperaturas Diárias')

plt.xlabel('Dias')

plt.ylabel('Temperatura (°C)')

plt.show()
```

## Pre-processamento de dados:

```python
# Normalizando os dados para o intervalo [0, 1]
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df)

# Função para criar sequências de dados
def create_sequences(data, seq_length):
    X = []
    y = []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

# Definindo o comprimento da sequência (30 dias)
seq_length = 30

# Criando as sequências de dados
X, y = create_sequences(scaled_data, seq_length)

# Dividindo os dados em conjuntos de treino e teste
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# Verificando as formas dos conjuntos de dados
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

### Na construção do modelo LSTM   foi utilizado o o algotirimo otimizador ADAM por conta de sua eficiência computacional e baixo consumo de memória,
### a função de perda escolhida foi o MSE por conta de ser uma boa escolha no processo de trabalho com dados de series temporais.
### o imput esta definido para 100, com saida para 1, com output sendo a temperatura prevista para o dia seguintes após a analise dos ultimos 30 dias.
### o treinamento do modelo foi definido em 42 épocas.

```python
# Construindo o modelo LSTM
model = Sequential()
model.add(Input(shape=(seq_length, 1)))  # Definindo a forma de entrada explicitamente
model.add(LSTM(100, activation='relu'))  # Camada LSTM com 50 unidades e ativação ReLU
model.add(Dense(1))  # Camada densa para a saída
model.compile(optimizer='adam', loss='mse')  # Compilando o modelo com otimizador Adam e função de perda MSE

# Resumo do modelo
model.summary()

# Treinando o modelo
history = model.fit(
    X_train, y_train, 
    epochs=42, 
    validation_data=(X_test, y_test), 
    verbose=1  # Exibe o progresso do treinamento
)

# Visualizando a perda durante o treinamento
plt.plot(history.history['loss'], label='Perda de treino') 
plt.plot(history.history['val_loss'], label='Perda de Validação')
plt.title('Perda do Modelo')
plt.xlabel('Épocas')
plt.ylabel('Perdas')
plt.legend()
plt.show()
```

# Previsão da temperatura com base nas ultimas 30 amostras de temperaturas medias diárias.

```python
# 1. 30 dias de dados
ultimos_30_dias = scaled_data[-30:]  # Pega os últimos 30 dias dos dados normalizados

# 2. Formate os dados para o modelo (formato: (1, 30, 1))
ultimos_30_dias = ultimos_30_dias.reshape((1, 30, 1))

# 3. previsão
temperatura_predita_normalizada = model.predict(ultimos_30_dias)

# 4.a normalização para obter o valor real
temperatura_predita = scaler.inverse_transform(temperatura_predita_normalizada)

# 5. resultado
print(f"Previsão da temperatura para o dia seguinte: {temperatura_predita[0][0]:.2f} °C")
```

## plotagem da verificação da temperatura dos ultimos 30 dias e previsão

```python
import matplotlib.pyplot as plt

# 1. 30 dias de dados
ultimos_30_dias = scaled_data[-30:]  # Pega os últimos 30 dias dos dados normalizados

# 2. dados para o modelo (formato: (1, 30, 1))
ultimos_30_dias_reshaped = ultimos_30_dias.reshape((1, 30, 1))

# 3. previsão para o dia seguinte
temperatura_predita_normalizada = model.predict(ultimos_30_dias_reshaped)
temperatura_predita = scaler.inverse_transform(temperatura_predita_normalizada)

# 4. Inverta a normalização dos últimos 30 dias para obter os valores reais
ultimos_30_dias_reais = scaler.inverse_transform(ultimos_30_dias.reshape(-1, 1))

# 5. Preparação os dados para o gráfico
dias = np.arange(1, 31)  # Dias 1 a 30
dias_com_previsao = np.arange(1, 32)  # Dias 1 a 31 (incluindo a previsão)

# Temperaturas dos últimos 30 dias
temperaturas_reais = ultimos_30_dias_reais.flatten()

# Temperatura prevista para o dia 31
temperatura_prevista = temperatura_predita[0][0]

# 6. Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(dias, temperaturas_reais, label='Temperaturas Reais (Últimos 30 Dias)', marker='o', color='blue')
plt.plot(dias_com_previsao[-1], temperatura_prevista, label='Previsão para o Dia 31', marker='x', color='red', markersize=10)
plt.title('Temperaturas dos Últimos 30 Dias e Previsão para o Dia Seguinte')
plt.xlabel('Dias')
plt.ylabel('Temperatura (°C)')
plt.xticks(dias_com_previsao)  # Mostra todos os dias no eixo X
plt.legend()
plt.grid(True)
plt.show()
```

# Validação e testagem do modelo.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Fazendo previsões nos dados de teste
y_pred = model.predict(X_test)

# Revertendo a normalização para obter as previsões na escala original
y_pred_actual = scaler.inverse_transform(y_pred)
y_test_actual = scaler.inverse_transform(y_test)

# Calculando as métricas
mse = mean_squared_error(y_test_actual, y_pred_actual)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test_actual, y_pred_actual)
r2 = r2_score(y_test_actual, y_pred_actual)

print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.2f}")
```
